# **Q.1 Writing two functions of given prototype in a file named activation.py**

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# (start, stop, no. of samples to generate)
x = np.linspace(-10, 10, 1000)
z = 1/(1 + np.exp(-x))
  
plt.plot(x, z)
plt.xlabel("z",fontsize=15)
plt.ylabel("Sigmoid(z)",fontsize=15)
plt.title("Graphs of Sigmoid(z) vs. z")
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
y = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

plt.plot(x,y)
plt.xlabel("z",fontsize=15)
plt.ylabel("tanh(z)",fontsize=15)
plt.title("Graph of tanh(z) vs. z",fontsize=20)
plt.show()
```

## **Below 2 graphs are in fact the overall graphs of Global Scores vs. Countries**.

```python
z = df['Global Score 2021']
x = df['Country Name']

plt.plot(x,z)
plt.xlabel("Country Name",fontsize=15)
plt.ylabel("Global Score 2021",fontsize=15)
plt.title("Global Score 2021",fontsize=20)
plt.show()
```

```python
y = df['Global Score 2019']
x = df['Country Name']
# Rotation for Country Name
plt.plot(x,y)
plt.xlabel("Country Name",fontsize=15)
plt.ylabel("Global Score 2019",fontsize=15)
plt.title("Global Score 2019",fontsize=20)
plt.show()
```
